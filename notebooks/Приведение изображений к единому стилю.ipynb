```python
import os
from PIL import Image, ImageDraw
import matplotlib.pyplot as plt
```

```python
# Простой вывод изображения из массива
def img_show(arr):
    plt.figure(figsize=(20,10))
    plt.imshow(arr)
    plt.show()
```

```python
class_lists = [['Evgo', 'Женька'],
    ['my_in', 'Юра'],
    ['beavers_in', 'Бобер'],
    ['Аня', 'Аня'],
    ['Ира', 'Ира'],
]
```

```python
for class_l in class_lists:
    peth_in = '../data/input/' + class_l[0]
    peth_out = '../data/work/' + class_l[1]
    for index, img_name in enumerate(os.listdir(peth_in)):
        print(img_name)
        image = Image.open(peth_in + '/' + img_name)
        image = image.convert('RGB')
        image = image.resize((256, 256))
        image.save(peth_out + '/' + str(index) + '.jpeg')
#     img_show(image)
```
